```python
newnric = []
for s in df_demo['NRIC']:
  if ' ' in s:
    s = s.replace(' ', '')

  if s[0].isalpha() and s[1].isalpha():
    s = s[0]+s[2:]+s[1]
  newnric.append(s)
df_demo['New NRIC'] = pd.Series(newnric)
df_demo
```

```python
df_demo['New NRIC'].unique()
```

```python
gendercode = []
for i in df_demo['Gender']:
  if i =='F':
    code = 1
  else:
    code = 2
  gendercode.append(code)
df_demo['Coding - Gender'] = pd.Series(gendercode)
df_demo
```

```python
age_group = []
for i in df_demo['Age']:
  if i in range(0,10):
    group = 'G1'
  elif i in range(10,20):
    group = 'G2'
  elif i in range(20,30):
    group = 'G3'
  elif i in range(30,40):
    group = 'G4'
  elif i in range(40,50):
    group = 'G5'
  elif i in range(50,60):
    group = 'G6'
  elif i in range(60,70):
    group = 'G7'
  elif i in range(70,80):
    group = 'G8'
  elif i in range(80,90):
    group = 'G9'
  else:
    group = 'G0'
  age_group.append(group)
df_demo['Age Group'] = pd.Series(age_group)
df_demo
```

```python
studynum = []
count = {'G1':0, 'G2':0, 'G3':0, 'G4':0, 'G5':0, 'G6':0, 'G7':0, 'G8':0, 'G9':0, 'G0':0}
for i in df_demo['Age Group']:
  count[i]+=1
  val = i + "-" +str(count[i])
  studynum.append(val)
df_demo['Study Number'] = pd.Series(studynum)
df_demo
```

```python
#There are some missing records in ‘Extra information’ worksheet. List down the number of records without Demographics information.

formatted_nric = []
for s in df_extra['NRIC']:
  if ' ' in s:
    s = s.replace(' ', '')

  if s[0].isalpha() and s[1].isalpha():
    s = s[0]+s[2:]+s[1]
  formatted_nric.append(s)

num_extra = len(set(formatted_nric)) #number of unique NRIC in extra information worksheet
print(" There are " ,num_demo - num_extra, ' NRICs that are not included in the Extra Information Worksheet')

missing_records = set(df_demo['New NRIC'].unique()) - set(formatted_nric)
df_missing = pd.DataFrame(missing_records, columns = ['NRIC'], index = range(1,14))
df_missing
```

```python
#Pivot Table
#(Age Group, Gender, Marital Status)
table1 = pd.DataFrame(df_demo['Age Group'].value_counts()).reset_index()
table1.columns = ['Value', 'Total']
table1 = table1.sort_values(by = ['Value'],ascending= True).reset_index(drop=True)
table1.loc["Age Group Total"] = ['',table1['Total'].sum()]

table2 = pd.DataFrame(df_demo['Gender'].value_counts()).reset_index()
table2.columns = ['Value', 'Total']
table2.loc["Gender Total"] = ['',table2['Total'].sum()]

table3 = pd.DataFrame(df_demo['Marital Status'].value_counts()).reset_index()
table3.columns = ['Value','Total']
table3.sort_values(by=['Value']).reset_index()
table3.loc["Marital Status Total"] = ['',table3['Total'].sum()]


```

```python
pivot = pd.concat([table1, table2, table3], axis=0, keys = ['Age Group', 'Gender', 'Marital Status'])
pivot
```

```python
pivot.to_excel('pivot_sheet.xlsx')
files.download('pivot_sheet.xlsx')
```
